```python
# Load the play-by-play data for a single season (e.g., 2023) to explore relevant columns for missed extra points
pbp_2023 = nflreadpy.load_pbp(seasons=[2023])

# Show the column names and sample rows to identify missed XP fields and join keys
display(pbp_2023.columns)
display(pbp_2023.head(10))
```

```python
# Aggregate missed extra points (missed or blocked) in 2023 season PBP using Polars syntax
import polars as pl

# Ensure pbp_2023 is a Polars DataFrame
def ensure_polars(df):
    if not isinstance(df, pl.DataFrame):
        return pl.from_pandas(df)
    return df

pbp_2023 = ensure_polars(pbp_2023)

# Filter for extra point attempts
xp_pbp = pbp_2023.filter(pl.col('extra_point_attempt') == 1)
# Filter for missed or blocked extra points
missed_xp = xp_pbp.filter(pl.col('extra_point_result').is_in(['missed', 'blocked']))
# Count missed XP per game_id
missed_xp_by_game = (
    missed_xp.group_by('game_id')
    .count()
    .select(['game_id', 'count'])
    .rename({'count': 'missed_extra_points'})
)
# Preview the output
display(missed_xp_by_game.head(10))
```

```python
# Aggregate missed extra points for each season batch (to avoid out-of-memory)
import polars as pl
import nflreadpy

def batch_aggregate_missed_xp(years):
    pbp = nflreadpy.load_pbp(seasons=years)
    pbp = pl.from_pandas(pbp) if not isinstance(pbp, pl.DataFrame) else pbp
    exps = pbp.filter(pl.col('extra_point_attempt') == 1)
    missed = exps.filter(pl.col('extra_point_result').is_in(['missed', 'blocked']))
    missed_games = (
        missed.group_by('game_id').len()
        .select(['game_id', 'len'])
        .rename({'len': 'missed_extra_points'})
    )
    return missed_games

# Start with the first decade as a test window
missed_xp_99_05 = batch_aggregate_missed_xp(list(range(1999, 2006)))
display(missed_xp_99_05.head(10))
```

```python
# Process and aggregate missed XP for all years in batches, then join to games and compute remaining fields
import polars as pl
from functools import reduce

# Window ranges to avoid OOM; adjust as safe
batches = [
    range(1999, 2006),  # 7 years
    range(2006, 2012),  # 6 years
    range(2012, 2018),  # 6 years
    range(2018, 2023),  # 5 years
    range(2023, 2026)   # 3 years
]

results = []
for years in batches:
    df = batch_aggregate_missed_xp(list(years))
    results.append(df)

# Concatenate all batches for missed XP per game
totals = pl.concat(results)

# Check if games is already polars; only convert if needed
if not isinstance(games, pl.DataFrame):
    games_pl = pl.from_pandas(games)
else:
    games_pl = games

# Left join: pull in missed_extra_points (defaults to 0 where not found)
games_with_xp = games_pl.join(
    totals,
    on='game_id',
    how='left'
).with_columns(
    pl.col('missed_extra_points').fill_null(0).cast(pl.Int32)
)

# Compute losing team, point differential (absolute value), etc.
final = games_with_xp.with_columns([
    pl.when(pl.col('home_score') > pl.col('away_score'))
      .then(pl.col('away_team'))
      .otherwise(pl.col('home_team'))
      .alias('losing_team'),
    (pl.col('home_score') - pl.col('away_score')).abs().alias('point_differential')
])

# Select desired columns
output_table = final.select([
    'game_id','season','week','home_team','away_team','losing_team',
    'home_score','away_score','point_differential','missed_extra_points'
])

# Show a preview of the finished table
output_table.head(10)
```

```python
# Filter table: missed_extra_points > 1 and point_differential <= missed_extra_points
filtered_games = output_table.filter(
    (pl.col('missed_extra_points') > 1) & (pl.col('point_differential') <= pl.col('missed_extra_points'))
)
# Show first 20 results for sanity check
filtered_games.head(20)
```

```python
# Aggregate failed extra points (corrected: extra_point_result == 'failed'), then redo join and filtering
import polars as pl

# Helper (corrected logic)
def batch_aggregate_failed_xp(years):
    pbp = nflreadpy.load_pbp(seasons=years)
    pbp = pl.from_pandas(pbp) if not isinstance(pbp, pl.DataFrame) else pbp
    exps = pbp.filter(pl.col('extra_point_attempt') == 1)
    failed = exps.filter(pl.col('extra_point_result') == 'failed')
    failed_games = (
        failed.group_by('game_id').len()
        .select(['game_id', 'len'])
        .rename({'len': 'failed_extra_points'})
    )
    return failed_games

# Re-run for all batches
failed_results = []
for years in batches:
    df = batch_aggregate_failed_xp(list(years))
    failed_results.append(df)

# Combine
failed_totals = pl.concat(failed_results)

# Rejoin to game schedule
final_failed = games_pl.join(
    failed_totals,
    on='game_id',
    how='left'
).with_columns(
    pl.col('failed_extra_points').fill_null(0).cast(pl.Int32)
)

# Add losing team, point diff again
final_failed = final_failed.with_columns([
    pl.when(pl.col('home_score') > pl.col('away_score'))
      .then(pl.col('away_team'))
      .otherwise(pl.col('home_team'))
      .alias('losing_team'),
    (pl.col('home_score') - pl.col('away_score')).abs().alias('point_differential')
])

failed_output_table = final_failed.select([
    'game_id','season','week','home_team','away_team','losing_team',
    'home_score','away_score','point_differential','failed_extra_points'
])

# Apply your filter: failed_extra_points > 1 and point_differential <= failed_extra_points
filtered_failed_games = failed_output_table.filter(
    (pl.col('failed_extra_points') > 1) & (pl.col('point_differential') <= pl.col('failed_extra_points'))
)

# Show filtered results and specifically check Giants vs Broncos 2025
gb2025 = failed_output_table.filter(
    (pl.col('season') == 2025) & (
        ((pl.col('home_team') == 'NYG') & (pl.col('away_team') == 'DEN')) |
        ((pl.col('home_team') == 'DEN') & (pl.col('away_team') == 'NYG'))
    )
)
filtered_failed_games.head(10), gb2025
```

```python
# Function to aggregate failed extra points by LOSING team for each game and batch (copied here for proper scope)
def batch_aggregate_failed_xp_loser(years, games_pl):
    import polars as pl
    import nflreadpy
    pbp = nflreadpy.load_pbp(seasons=years)
    pbp = pl.from_pandas(pbp) if not isinstance(pbp, pl.DataFrame) else pbp
    # Join game table to get losing_team for each game_id
    games_slice = games_pl.filter(pl.col('season').is_in(years))
    meta_cols = ['game_id','home_team','away_team','home_score','away_score']
    games_slice = games_slice.select(meta_cols)
    games_slice = games_slice.with_columns([
        pl.when(pl.col('home_score') > pl.col('away_score'))
          .then(pl.col('away_team'))
          .otherwise(pl.col('home_team'))
          .alias('losing_team')
    ])
    pbp_merged = pbp.join(games_slice.select(['game_id','losing_team']), on='game_id', how='left')
    failed = pbp_merged.filter(
        (pl.col('extra_point_attempt') == 1) &
        (pl.col('extra_point_result') == 'failed') &
        (pl.col('posteam') == pl.col('losing_team'))
    )
    failed_games_loser = (
        failed.group_by('game_id').len()
        .select(['game_id', 'len'])
        .rename({'len': 'failed_extra_points_loser'})
    )
    return failed_games_loser

batched_failed_loser = []
for years in batches:
    batch_df = batch_aggregate_failed_xp_loser(list(years), games_pl)
    batched_failed_loser.append(batch_df)

failed_loser_totals = pl.concat(batched_failed_loser)

# Join to game table
final_loser_tbl = games_pl.join(
    failed_loser_totals,
    on='game_id',
    how='left'
).with_columns(
    pl.col('failed_extra_points_loser').fill_null(0).cast(pl.Int32)
)

# Add losing team, point diff columns
final_loser_tbl = final_loser_tbl.with_columns([
    pl.when(pl.col('home_score') > pl.col('away_score'))
      .then(pl.col('away_team'))
      .otherwise(pl.col('home_team'))
      .alias('losing_team'),
    (pl.col('home_score') - pl.col('away_score')).abs().alias('point_differential')
])

loser_output_table = final_loser_tbl.select([
    'game_id','season','week','home_team','away_team','losing_team',
    'home_score','away_score','point_differential','failed_extra_points_loser'
])

# Show a preview
display(loser_output_table.head(10))
```
